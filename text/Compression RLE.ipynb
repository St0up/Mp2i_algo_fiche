# Compression Run-length Encoding (RLE)

Compresser une liste de caractères par le nombre d'éléments pareil à la suite.

exemple: compresser "aaaabccbbb" en  
[('a', 4); ('b', 1); ('c', 2); ('b', 3)]

```ocaml
let rle_code s = 
    let ind = ref s.[0] in
    let nb = ref 1 in
    let l = ref [] in
    for i = 1 to String.length s -1 do
        if !ind <> s.[i] then (
            l := (!ind,!nb)::!l;
            ind := s.[i];
            nb := 1
        )
        else incr nb
    done;
    List.rev ((!ind,!nb)::!l)
```

```ocaml
let rec rle_decode = function
    |[] -> ""
    |(c,n)::q -> (String.make n c)^rle_decode q 
```
