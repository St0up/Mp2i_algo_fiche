# Parcours d'arbres

Le parcours en profondeur va le plus profondément possible avant de passer à la prochaine.

```ocaml
type 'a binary_tree = 
    | E
    | N of 'a * 'a binary_tree * 'a binary_tree;;
```

### Parcours préfixe

D'abord r, puis les noeuds de g (appel récursif), puis les noeuds
de d (appel récursif)

```ocaml
let rec prefixe acc a = match a with
    | E -> acc
    | N(r,g,d) -> r::prefixe (prefixe acc d) g;;
```

### Parcours infixe

D'abord les noeuds de g (appel récursif), puis r, puis les noeuds de d (appel récursif)

```ocaml
let rec infixe acc a = match a with
    | E -> acc
    | N(r,g,d) -> infixe (r::(infixe acc d)) g;;
```

### Parcours suffixe

D'abord les noeuds de g (appel récursif), puis les noeuds de d (appel récursif), puis r

```ocaml
let rec suffixe acc a = match a with
    | E -> acc
    | N(r,g,d) -> suffixe ((suffixe acc d)@r) g;;
```
