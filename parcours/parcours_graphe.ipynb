# Parcours dans les graphes

```ocaml
type 'a graphe = {n : int; adj : 'a array}
```

### Parcours en profondeur (DFS)

On visite les sommets le plus profondément possible avant de revenir en arrière

```ocaml
let dfs g r =
    let seen = Array.make g.n true in
    let rec aux v = 
        if seen.(v) then(
            seen.(v) <- false;
            List.iter aux g.adj.(v)
        )in
    aux r;;
```

### Parcours en largeur (BFS)

On visite les sommets par distance croissante depuis une racine

```ocaml
let bfs g r =
    let dist = Array.make g.n (-1) in
    let rec aux d cur next = match cur with
        | [] -> if next <> [] then aux (d+1) next []
        | u::q when dist.(u) <> -1 -> aux d q next
        | u::q -> dist.(u) <- d;
                    aux d q ((g.adj.(u))@ next) in
    aux 0 [r] [];
    dist
```
