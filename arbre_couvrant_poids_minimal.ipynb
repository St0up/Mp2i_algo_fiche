# Arbre couvrant de poids minimal

## Algorithme de Kruskal

Trier les arêtes par poids croissant  
Commencer avec un arbre T vide (aucune arête).  

Pour chaque arête e par poids croissant:  
&nbsp; &nbsp; &nbsp; Si l'ajout de e ne créé pas de cycle dans T:  
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; Ajouter e à T

```ocaml
type ('v, 'w) graph = {
    n: int;
    add_edge: 'v -> 'v -> 'w -> unit;
    del_edge: 'v -> 'v -> unit;
    edges: unit -> ('v*'v) list;
    w : 'v -> 'v -> 'w option;
    adj: 'v -> 'v list
}
```

```ocaml
let kruskal g = 
    let uf = create () in
    let rec aux = function
        |[] -> []
        |(u, v)::q when find uf u = find uf v -> aux q
        |(u, v)::q -> (union uf u v; (u, v)::aux q) in
    g.edges ()
    |> List.sort (fun e f -> g.w e - g.w f)
    |> aux
```

## Algorithme de Prim

Commencer avec un arbre T contenant une seul sommet.  

Tant que T ne contient pas tous les sommets:  
&nbsp; &nbsp; &nbsp; Ajouter l'arête sortante de T de poids minimum

Pour implémenter Prim, on utilise une file de priorité pq :  
* pq contient les arêtes sortantes de T
* les éléments de pq sont ordonnés par poids croissants :  
&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; take_min pq renvoie donc l'arête sortante de T de poids minimum
